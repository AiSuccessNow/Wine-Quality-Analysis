## Creating a visualization to display light-bodied, medium bodied and full-bodied red wine using pH and alcohol content 

```python
#Isolating alcohol and pH in the red wine dataframe 
df = red_wine_df[["alcohol", "pH"]]
df
```

```python
#highlighting different pH ranges based on the body of wine 
def body(x):
    if x <12.5:
        return "pink"
    elif x>13.5:
        return "darkred"
    else: 
        return "crimson"
```

```python
#testing the code
body(12.4)
```

```python
#Applying the body function to the pH levels
df["Body"] = df["alcohol"].apply(body)
df.head()
```

```python
#Creating a visualization of the different bodies of red wine based on pH and alcohol content
red_wine_df['color'] = red_wine_df['alcohol'].apply(body)
# Plot the scatter plot with colors based on the new column
plt.figure(figsize=(12, 8))
sns.scatterplot(data=red_wine_df, x="alcohol", y="pH", hue='color', palette=['pink', 'darkred', 'crimson'], legend=False)
plt.title("Red Wine Bodies Based on Alcohol and pH Values")
# Perform linear regression
x_values = red_wine_df['alcohol']
y_values = red_wine_df['pH']
# Add a custom legend
legend_labels = {
    'pink': 'Light-Bodied',
    'darkred': 'Medium-Bodied',
    'crimson': 'Full-Bodied'
}
handles = [mpatches.Patch(color=color, label=label) for color, label in legend_labels.items()]
plt.legend(handles=handles, title='Alcohol Content')
plt.xlabel("Alcohol")
plt.ylabel("pH")
```

## Visualization of the different bodies of white wine with a line of best fit

```python
import matplotlib.patches as mpatches
# Function to determine color based on alcohol content
def white_wine_body(x):
    if x < 12.5:
        return "paleturquoise"
    elif x > 13.5:
        return "deepskyblue"
    else:
        return "darkblue"
# Apply the function to create a new column for colors
df_white_wine['color'] = df_white_wine['alcohol'].apply(white_wine_body)
# Plot the scatter plot with colors based on the new column
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_white_wine, x="alcohol", y="pH", hue='color', palette=['paleturquoise', 'darkblue', 'deepskyblue'], legend=False)
plt.title("White Wine Alcohol vs pH")
# Perform linear regression
x_values = df_white_wine['alcohol']
y_values = df_white_wine['pH']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = f"y = {slope:.2f}x + {intercept:.2f}"
# Plot the line of best fit
plt.plot(x_values, regress_values, color='red')
# Add textbox with the regression equation
plt.text(0.05, 0.95, line_eq, transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', edgecolor='black', facecolor='white'))
# Add a custom legend
legend_labels = {
    'paleturquoise': 'Light-Bodied',
    'darkblue': 'Medium-Bodied',
    'deepskyblue': 'Full-Bodied'
}
handles = [mpatches.Patch(color=color, label=label) for color, label in legend_labels.items()]
plt.legend(handles=handles, title='Alcohol Content')
plt.xlabel("Alcohol")
plt.ylabel("pH")
plt.show()
```

## Illustrating the relationship between pH and residual sugar in white wine with a line of best fit

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=white_wine_df, x="pH", y="residual sugar")
plt.title("White Wine pH vs Residual Sugar")
x_values = white_wine_df['pH']
y_values = white_wine_df['residual sugar']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))
# Plot the line of best fit
plt.plot(x_values, regress_values, color='red', label=line_eq)
plt.legend()
plt.show()
```
