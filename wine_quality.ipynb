```python
red_wine_df = pd.read_csv('/Users/ivanvargas/Documents/AI_Bootcamp/Wine-Quality-Analysis/wine_quality_red.csv', delimiter=';')
white_wine_df = pd.read_csv('/Users/ivanvargas/Documents/AI_Bootcamp/Wine-Quality-Analysis/wine_quality_white.csv', delimiter=';')
```

```python
combined_wine_df = pd.concat([red_wine_df, white_wine_df], ignore_index=True, sort=False)

combined_wine_df
```

```python
combined_wine_df = combined_wine_df.sort_values(by=['quality'], ascending=False)

combined_wine_df.head(50)
```

```python
plt.figure(figsize=(12, 6))
for quality in sorted(red_wine_df['quality'].unique()):
    subset = red_wine_df[red_wine_df['quality'] == quality]
    plt.hist(subset['alcohol'], bins=15, alpha=0.5, label=f'Quality {quality}')
plt.title('Distribution of Alcohol Content by Red Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.legend(title='Quality Score')
plt.grid(axis='y')
plt.show()
```

```python
chemical_properties = combined_wine_df.drop(columns=['color'])
```

```python
wine_correlation = chemical_properties.corr()
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(wine_correlation, cmap='coolwarm', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(wine_correlation)), wine_correlation.columns, rotation=90)
plt.yticks(range(len(wine_correlation)), wine_correlation.columns)
plt.title('Correlation Heatmap of Wine Chemical Properties')
plt.show()
```
