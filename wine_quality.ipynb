```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load red wine and white wine datasets
red_wine_df = pd.read_csv('/Users/ivanvargas/Documents/AI_Bootcamp/Wine-Quality-Analysis/wine_quality_red.csv', delimiter=';')
white_wine_df = pd.read_csv('/Users/ivanvargas/Documents/AI_Bootcamp/Wine-Quality-Analysis/wine_quality_white.csv', delimiter=';')
```

```python
# Display the red wine dataset
red_wine_df
```

```python
# Add a color column to identify red wines before concatenation
red_wine_df['color'] = 'red'
```

```python
red_wine_df
```

```python
# Get a concise summary of the red wine DataFrame
red_wine_df.info()
```

```python
# Display the columns of the red wine DataFrame
red_wine_df.columns
```

```python
# Display the white wine dataset
white_wine_df
```

```python
# Add a color column to identify white wines before concatenation
white_wine_df['color'] = 'white'
```

```python
white_wine_df
```

```python
# Get a concise summary of the white wine DataFrame
white_wine_df.info()
```

```python
# Display the columns of the white wine DataFrame
white_wine_df.columns
```

```python
# Combine red and white wine DataFrames into one
combined_wine_df = pd.concat([red_wine_df, white_wine_df], ignore_index=True, sort=False)
# Display the combined wine DataFrame
combined_wine_df
```

```python
# Sort the combined DataFrame by quality in descending order
combined_wine_df = combined_wine_df.sort_values(by=['quality'], ascending=False)
# Display the combined wine DataFrame
combined_wine_df
```

```python
# Save the cleaned combined DataFrame to a CSV file
combined_wine_df.to_csv('cleaned_wine_data.csv', index=False)
```

Calculate the mean and range of 'pH' for red wine, then create a graph to show the distribution of pH content by wine quality

```python
plt.figure()
for quality in sorted(red_wine_df['quality'].unique()):
    subset = red_wine_df[red_wine_df['quality'] == quality]
    plt.hist(subset['pH'], bins=25, alpha=0.4, label=f'Quality {quality}')
plt.title('Distribution of PH Content by Red Wine Quality')
plt.xlabel('pH')
plt.ylabel('Frequency')
plt.legend(title='Quality Score')
plt.grid(axis='y')
plt.show()
```

Create graphs to show the distribution of alcohol content by wine quality for red wine, white wine, and the combined DataFrame

```python
plt.figure()
for quality in sorted(red_wine_df['quality'].unique()):
    subset = red_wine_df[red_wine_df['quality'] == quality]
    plt.hist(subset['alcohol'], bins=15, alpha=0.5, label=f'Quality {quality}')
plt.title('Distribution of Alcohol Content by Red Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.legend(title='Quality Score')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure()
for quality in sorted(white_wine_df['quality'].unique()):
    subset = white_wine_df[white_wine_df['quality'] == quality]
    plt.hist(subset['alcohol'], bins=15, alpha=0.5, label=f'Quality {quality}')
plt.title('Distribution of Alcohol Content by White Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.legend(title='Quality Score')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure()
for quality in sorted(combined_wine_df['quality'].unique()):
    subset = combined_wine_df[combined_wine_df['quality'] == quality]
    plt.hist(subset['alcohol'], bins=15, alpha=0.5, label=f'Quality {quality}')
plt.title('Distribution of Alcohol Content by Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Frequency')
plt.legend(title='Quality Score')
plt.grid(axis='y')
plt.show()
```
