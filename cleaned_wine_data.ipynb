```python
import pandas as pd
```

```python
red_wine_df = pd.read_csv(r"C:\Users\rasha\Group4\Group4\Wine-Quality-Analysis\winequality-red.csv", delimiter=";")
white_wine_df = pd.read_csv(r"C:\Users\rasha\Group4\Group4\Wine-Quality-Analysis\winequality-white.csv", delimiter=";")
```

```python
#concatenate the two dataframes
wine_df = pd.concat([red_wine_df, white_wine_df], ignore_index=True)
#save the concatenated dataframe to a csv file
wine_df.to_csv("C:\\Users\\rasha\\Group4\\winequality.csv", index=False)
#read the concatenated dataframe from the csv file
wine_df = pd.read_csv("C:\\Users\\rasha\\Group4\\winequality.csv")
```

```python
# Execute the cell that defines combined_wine_df before running this cell
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
# Plot the distribution of wine qualities
plt.figure(figsize=(10, 6))
sns.countplot(x='quality', data=wine_df, palette='viridis')
plt.title('Distribution of Wine Qualities')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
#The distribution plot shows the frequency of each wine quality rating in the dataset. The majority of wines have a quality rating around 6 to 8.
```

```python
# Calculate the average chlorides content for each quality rating
avg_chlorides_quality = wine_df.groupby('quality')['chlorides'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_chlorides_quality.index, avg_chlorides_quality.values, marker='o', linestyle='-', color='b')
plt.title('Average Chlorides Content vs Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Chlorides Content')
plt.grid(True)
plt.show()
```

```python
#The line chart shows the average chlorides content for each wine quality rating. It illustrates that wines with higher quality generally have lower average chlorides content, supporting the observation that lower chloride content is associated with higher quality wines.
```

```python
# Calculate the average sulphates content for each quality rating
avg_sulphates_quality = wine_df.groupby('quality')['sulphates'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_sulphates_quality.index, avg_sulphates_quality.values, marker='o', linestyle='-', color='r')
plt.title('Average Sulphates Content vs Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Sulphates Content')
plt.grid(True)
plt.show()
```

```python
#The line chart shows the average sulphates content for each wine quality rating. It indicates that higher quality wines generally have higher average sulphates content.
```

```python
# Calculate the average citric acid content for each quality rating
avg_citric_acid_quality = wine_df.groupby('quality')['citric acid'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_citric_acid_quality.index, avg_citric_acid_quality.values, marker='o', linestyle='-', color='g')
plt.title('Average Citric Acid Content vs Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Citric Acid Content')
plt.grid(True)
plt.show()
```

```python
#The line chart shows the average citric acid content for each wine quality rating. It indicates that higher quality wines tend to have slightly higher average citric acid content.
```

```python
# Calculate the average alcohol content for each quality rating
avg_alcohol_quality = wine_df.groupby('quality')['alcohol'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_alcohol_quality.index, y=avg_alcohol_quality.values, palette='viridis')
plt.title('Average Alcohol Content vs Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Average Alcohol Content (%)')
plt.show()
```

```python
#The bar plot shows the average alcohol content for each wine quality rating. It illustrates that higher quality wines tend to have higher average alcohol content. 
```

```python
#show me a chart that shows the total of each quality of wine
plt.figure(figsize=(10, 6))
wine_df['quality'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Total Count of Each Wine Quality')
plt.xlabel('Quality Score')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
#The bar chart shows the total count of each wine quality rating. It highlights the distribution of different quality ratings in the dataset, with most wines having quality ratings around 6 to 8.
```

```python
# Calculate the average volatile acidity for each quality rating
avg_volatile_acidity_quality = wine_df.groupby('quality')['volatile acidity'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_volatile_acidity_quality.index, avg_volatile_acidity_quality.values, marker='o', linestyle='-', color='orange')
plt.title('Average Volatile Acidity vs Wine Quality')
plt.xlabel('Wine Quality')
plt.ylabel('Average Volatile Acidity')
plt.grid(True)
plt.show()
```

```python
#The line chart shows the average volatile acidity for each wine quality rating. It indicates that higher quality wines tend to have lower average volatile acidity.
```
