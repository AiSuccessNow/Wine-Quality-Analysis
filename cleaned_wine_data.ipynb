```python
import pandas as pd
```

```python
red_wine_df = pd.read_csv(r"C:\Users\rasha\Group4\Group4\Wine-Quality-Analysis\winequality-red.csv", delimiter=";")
white_wine_df = pd.read_csv(r"C:\Users\rasha\Group4\Group4\Wine-Quality-Analysis\winequality-white.csv", delimiter=";")
```

```python
#concatenate the two dataframes
wine_df = pd.concat([red_wine_df, white_wine_df], ignore_index=True)
#save the concatenated dataframe to a csv file
wine_df.to_csv("C:\\Users\\rasha\\Group4\\winequality.csv", index=False)
#read the concatenated dataframe from the csv file
wine_df = pd.read_csv("C:\\Users\\rasha\\Group4\\winequality.csv")
```

### Red Wine Exploration

```python
red_wine_df.groupby('quality')[["sulphates"]].mean().plot(kind='line',color="green",figsize=(12, 4),fontsize=10)
plt.title('Average Sulphates by Red Wine Quality')
plt.xlabel('Wine Quality')
plt.legend(fontsize = 12)
plt.ylabel('Sulphates')
plt.grid(axis='y')
plt.show()
```

```python
red_wine_df['quality'].value_counts().sort_index().plot(kind='pie',autopct='%1.1f%%',cmap="RdYlGn",figsize=(12,12),fontsize=18)
plt.title('Red Wines Percentage of Volume Distribution Based on Quality Ranking')
plt.legend(title='Wine Quality',fontsize=15 )
plt.grid(axis='y')
plt.style.use("grayscale")
plt.show()
```

### White Wine Exploration

```python
white_wine_df.groupby('quality')[["sulphates"]].mean().plot(kind='line',color="blue",figsize=(12, 4),fontsize=10)
plt.title('Average Sulphates by White Wine Quality')
plt.xlabel('Wine Quality')
plt.legend(fontsize = 12)
plt.ylabel('Sulphates')
plt.grid(axis='y')
plt.show()
```

```python
white_wine_df['quality'].value_counts().sort_index().plot(kind='pie',autopct='%1.1f%%',cmap="RdYlGn",figsize=(12,12),fontsize=18)
plt.title('White Wines Percentage of Volume Distribution based on Quality Ranking')
plt.legend(title='Wine Quality',fontsize=10 )
plt.style.use("grayscale")
#plt.style.use("dark_background")
plt.grid(axis='y')
plt.show()
```
